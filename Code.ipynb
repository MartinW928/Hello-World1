```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock_data = pd.read_csv("/Users/martinwch/downloads/stock_data.csv", parse_dates=[0], index_col=[0]).dropna()
```

```python
benchmark_data = pd.read_csv("/Users/martinwch/downloads/benchmark_data.csv", parse_dates=[0], index_col=[0]).dropna()
```

```python
print('Stocks\n')
stock_data.info()
print(stock_data.head())
```

```python
print('\benchmark\n')
benchmark_data.info()
print(benchmark_data.head())
```

```python
stock_data.plot(title='Stock Data', subplots=True)
benchmark_data.describe()
```

```python
benchmark_data.plot(title='S&P 500')
```

```python
stock_returns = stock_data.pct_change()
stock_returns.plot()
stock_returns.describe()
```

```python
sp_returns = benchmark_data['S&P 500'].pct_change()
sp_returns.plot()
sp_returns.describe()
```

```python
excess_returns = stock_returns.sub(sp_returns, axis=0)
excess_returns.plot()
excess_returns.describe()
```

```python
avg_excess_return = excess_returns.mean()
avg_excess_return.plot.bar(title='Mean of the Return Difference')
```

```python
sd_excess_return = excess_returns.std()
sd_excess_return.plot.bar(title='Standard Deviation of the Return Difference')
```

```python
daily_sharp_ratio = avg_excess_return.div(sd_excess_return)
```

```python
n_days=252
annual_factor = np.sqrt(n_days)
```

```python
annual_sharpe_ratio = daily_sharp_ratio.mul(annual_factor)
```

```python
annual_sharpe_ratio.plot.bar(title='Annualized Sharpe Ratio: Stocks vs S&P500')
```
